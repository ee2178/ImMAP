```python
import torch
import torch.nn as nn
import torch.nn.functional as F

from visualization.image import plot_image
from operators import Mask, FFT2D, Sense
from physics.mask import make_acc_mask
from datasets.fastmri.common import load_knee_data
from solvers.tvd import tvd
```

```python
import torch
import numpy as np

from models.components import ST
from solvers.cg import cg
from solvers.rest import img_2_norm
from operators.difference import Diff2D, Diff2D_FFT

def tvd(y,      # Measurement
        E,      # Measurement Operator
        lam,    # Objective Function penalty weight
        rho,    # Augmented Lagrangian multiplier
        maxit = 100,
        tol = 1e-3,
        ):

    # Here, we implement Total Variation Denoising for solving general linear inverse problems through a scaled-form ADMM style decomposition.

    # First, initialize our difference operator.
    D = Diff2D_FFT()
    # Second, initialize our forward operator for CG
    def A(x, E=E, D=D, rho = rho):
        return E.normal(x)+rho*D.normal(x)

    # Zero initialization for z and v
    z = torch.zeros_like(E.H(y))
    v = torch.zeros_like(E.H(y))
    print(z.shape)
    print(v.shape)
    # Precompute EHy
    EHy = E.H(y)

    tol_reached=False
    for i in range(maxit):
        # x-update. Here, we solve a symmetric linear system, so use CG
        x, _ = cg(A = A, b = EHy+rho*D.H(v-z), max_iter = 100, tol = 1e-3)
        # z-update, soft thresholding.
        z = ST(D(x)-v, lam/rho)
        # v-update, i.e. dual ascent
        v = D(x)-z
        # Early exit check
        r = img_2_norm(D(x)-z)
        if r < tol:
            print(f"Reached tolerance at Iteration {i}")
            tol_reached=True
            return x, z, v, tol_reached

    return x, z, v, tol_reached
```

```python
device = 'cuda'
```

```python
# SET UP AN MRI RECON PROBLEM
y, smaps, knee_mask, gnd_truth = load_knee_data(slice_idx=15, device=device)
```

```python
mask = make_acc_mask(shape=y[0,0].shape, accel = 2, acs_lines=20, device = device)
```

```python
# Creating measurement operator
E = Mask(mask) @ FFT2D() @ Sense(smaps)
```

```python
# Solving via TVD
x, z, v, tol_reached = tvd(y,      
        E=E,      
        lam=2.0,    
        rho=0.1,    
        maxit = 100,
        tol = 1e-3)
```
