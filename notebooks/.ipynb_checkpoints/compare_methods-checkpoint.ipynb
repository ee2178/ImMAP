# Side-by-side: UNet vs image-domain I2SB vs latent I2SB

Runs all methods on the **same** batch and stacks them against the T1ce ground truth:
1. **UNet (yhat)** -- the e2e synthesis output.
2. **Image-domain I2SB** -- your chosen trained regressor (`IMAGE_I2SB_CFG`); x1 is auto-set to yhat
   or T1 depending on that run's `x1_source`.
3. **(optional) Latent I2SB** -- design-A pure-latent sampler (`RUN_LATENT`; currently weak).

**Prereqs (HPC):** the trained runs' saved `config.json` + `net.ckpt`, and the synth run. Edit the
`os.chdir` path. Latent is behind a flag so you can drop it.

```python
import os, json
import numpy as np
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline

os.chdir('/scratch/ee2178/ImMAP')     # <-- EDIT to your repo root

# ---- pick your runs (SAVED config.json paths) ----
SYNTH_CFG       = "trained_nets/brats/Synth_T1ce_Pretrain_VGG_CosLR/config.json"   # the UNet -> yhat
IMAGE_I2SB_CFG  = "trained_nets/brats/I2SB_T1ce_from_Synth/config.json"            # <-- CHOOSE the image i2sb regressor
LATENT_I2SB_CFG = "trained_nets/brats/Latent_I2SB_T1ce/config.json"               # latent run
RUN_LATENT      = True          # set False to skip the (currently poor) latent method
SPLIT, NFE, USE_EMA = "val", 100, True
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

from models import build_model
from training.common import load_model, apply_loss_mask
from training.metrics import compute_metrics
from physics.bbridge import build_bridge
from diffusion.i2sb import i2sb_sample
import datasets                                   # registers loaders
from datasets.registry import build_loader

# build_model + load net.ckpt (+ EMA shadow) + compile flex; returns (net, cfg)
def load_net_ema(config_path, use_ema=USE_EMA):
    with open(config_path) as f:
        c = json.load(f)
    sd = os.path.dirname(config_path)
    net = build_model(c).to(device)
    net.load_state_dict(torch.load(os.path.join(sd, "net.ckpt"), map_location=device,
                                   weights_only=False)["model_state_dict"])
    net.eval()
    ema_p = os.path.join(sd, "ema.pt")
    if use_ema and os.path.exists(ema_p):
        shadow = torch.load(ema_p, map_location=device, weights_only=False)["shadow"]
        for p, s in zip([p for p in net.parameters() if p.requires_grad], shadow):
            p.data.copy_(s.to(p.device))
    if getattr(net, "attn_backend", None) == "flex":
        net.compile_flex()
    return net, c

print("device:", device)
```

## 1. One shared batch + the UNet output (yhat)

A single canonical loader (x1=T1, cond=[FLAIR,T1,T2], center-cropped 192) gives the batch every method sees. `yhat` is computed from the synth net on `cond`, so it doesn't depend on the precompute.

```python
# canonical loader: x1 = T1, cond = [FLAIR,T1,T2], fixed 192 crop, from the image-i2sb run's dataset
img_cfg = json.load(open(IMAGE_I2SB_CFG))
base = dict(img_cfg["data"][SPLIT])
base.update(name="i2sb", x1_source="contrast", cond_idx=[0, 1, 3], center_crop=192,
            random_flips=False, num_workers=0)
base.pop("crop_size", None)
loader = build_loader(base, shuffle=True, drop_last=False)
x0, x1_t1, cond, mask = next(iter(loader))          # x0=T1ce GT, x1_t1=T1 prior, cond=[FLAIR,T1,T2]
x0, x1_t1, mask, cond = x0.to(device), x1_t1.to(device), mask.to(device), cond.to(device)

# yhat = UNet(cond): undo the i2sb /scales, run the synth net (pad to 2^num_pool), re-apply x0 scale
synth = load_model(SYNTH_CFG, device=device); synth.eval()
scfg = json.load(open(SYNTH_CFG))
scales = torch.tensor(base["scales"], device=device)
cidx, x0i = list(base["cond_idx"]), int(base["x0_idx"])
mult = 2 ** int(scfg["model"]["params"].get("num_pool_layers", 5))
assert list(scfg["data"]["train"]["input_idx"]) == cidx, "synth input_idx != cond_idx"
with torch.no_grad():
    si = cond * scales[cidx].view(1, -1, 1, 1)
    H, W = si.shape[-2:]; ph, pw = (-H) % mult, (-W) % mult
    if ph or pw:
        si = F.pad(si, (0, pw, 0, ph), mode="replicate")
    yhat = synth(si)[..., :H, :W] / scales[x0i]
print("batch:", tuple(x0.shape), "| yhat:", tuple(yhat.shape))
```

## 2. Image-domain I2SB (your chosen regressor)

Loads `IMAGE_I2SB_CFG` and samples with the standard `i2sb_sample`. `x1` is set to `yhat` when that run's `x1_source=="synth"`, else T1; conditioning is passed only if the model's C > 1.

```python
net_img, cfg_img = load_net_ema(IMAGE_I2SB_CFG)
ici = cfg_img["i2sb"]
bridge_img = build_bridge(bridge_type=ici.get("bridge_type", "brownian"), n_points=ici.get("n_points", 1000),
                          device=device, tau=ici.get("tau", 0.19), shape=ici.get("bridge_shape", "constant"),
                          beta_max=ici.get("beta_max", 0.3))
x1_img = yhat if cfg_img["data"]["train"].get("x1_source") == "synth" else x1_t1
cond_img = cond if int(cfg_img["model"]["params"]["C"]) > 1 else None
recon_img, _, _ = i2sb_sample(net_img, x1_img, bridge_img, cond=cond_img, nfe=NFE,
                              ot_ode=ici.get("ot_ode", False), parameterization=ici.get("parameterization", "x0"),
                              clip_denoise=ici.get("clip_denoise", False),
                              target_channels=ici.get("target_channels", 1), verbose=True)
print("image I2SB done  (x1 =", "yhat" if x1_img is yhat else "T1", "| tau =", ici.get("tau"), ")")
```

## 3. (optional) Latent I2SB

Design-A pure-latent sampler: encode the prior, reverse in latent, decode once. Skipped if `RUN_LATENT=False`.

```python
recon_lat = None
if RUN_LATENT:
    from training.latent_i2sb import _load_frozen_dict, latent_i2sb_sample
    R, cfg_lat = load_net_ema(LATENT_I2SB_CFG)
    li, ld = cfg_lat["i2sb"], cfg_lat["dicts"]
    D_joint = _load_frozen_dict(ld["joint_dict_config"], device, backend=ld.get("dict_backend"))
    D_t1ce  = _load_frozen_dict(ld["t1ce_dict_config"],  device, backend=ld.get("dict_backend"))
    bridge_lat = build_bridge(bridge_type=li.get("bridge_type", "brownian"), n_points=li.get("n_points", 1000),
                              device=device, tau=li.get("tau", 0.024), shape=li.get("bridge_shape", "constant"),
                              beta_max=li.get("beta_max", 0.3))
    recon_lat = latent_i2sb_sample(D_joint, R, D_t1ce, x1_t1, cond, bridge_lat,
                                   conditional=li.get("conditional", True), M=D_joint.M, nfe=NFE,
                                   ot_ode=li.get("ot_ode", False), decode_dc=li.get("decode_dc", "x1_mean"))
    print("latent I2SB done  (tau =", li.get("tau"), ")")
else:
    print("latent I2SB skipped")
```

## 4. Side by side

T1 prior | UNet (yhat) | image I2SB | (latent I2SB) | T1ce GT, with masked PSNR under each. Watch whether each stage actually improves on the one before.

```python
def _win(*imgs, m):
    v = torch.cat([im[m.bool().expand_as(im)].flatten() for im in imgs]).float().cpu()
    return float(v.quantile(0.01)), float(v.quantile(0.99))

methods = [("T1 prior", x1_t1), ("UNet (yhat)", yhat), ("image I2SB", recon_img)]
if recon_lat is not None:
    methods.append(("latent I2SB", recon_lat))
methods.append(("T1ce GT", x0))

nshow = min(4, x0.shape[0])
vmn, vmx = _win(x1_t1[:nshow], x0[:nshow], m=mask[:nshow])
ncol = len(methods)
fig, ax = plt.subplots(nshow, ncol, figsize=(2.2 * ncol, 2.5 * nshow)); ax = np.atleast_2d(ax)
for i in range(nshow):
    for j, (name, img) in enumerate(methods):
        g, r = apply_loss_mask(x0[i:i + 1], img[i:i + 1], mask[i:i + 1], True)
        ttl = name if name == "T1ce GT" else f"{name}\n{float(compute_metrics(g, r, psnr_only=True)['psnr']):.1f} dB"
        ax[i, j].imshow(r[0, 0].detach().cpu().numpy(), cmap="gray", vmin=vmn, vmax=vmx)
        ax[i, j].set_title(ttl, fontsize=8); ax[i, j].set_xticks([]); ax[i, j].set_yticks([])
plt.tight_layout(); plt.show()
```

## 5. (optional) Aggregate metrics over a few batches

Re-runs each method on `N_BATCHES` fresh batches (the samplers are the cost -- keep it small) and reports mean masked PSNR / SSIM per method.

```python
N_BATCHES = 3

def _yhat(cb):
    with torch.no_grad():
        s = cb * scales[cidx].view(1, -1, 1, 1)
        h, w = s.shape[-2:]; a, b = (-h) % mult, (-w) % mult
        if a or b:
            s = F.pad(s, (0, b, 0, a), mode="replicate")
        return synth(s)[..., :h, :w] / scales[x0i]

def _img_i2sb(x1t1, yh, cb):
    x1b = yh if cfg_img["data"]["train"].get("x1_source") == "synth" else x1t1
    cbb = cb if int(cfg_img["model"]["params"]["C"]) > 1 else None
    r, _, _ = i2sb_sample(net_img, x1b, bridge_img, cond=cbb, nfe=NFE, ot_ode=ici.get("ot_ode", False),
                          parameterization=ici.get("parameterization", "x0"),
                          clip_denoise=ici.get("clip_denoise", False),
                          target_channels=ici.get("target_channels", 1), verbose=False)
    return r

agg = {"UNet (yhat)": [], "image I2SB": []}
if recon_lat is not None:
    agg["latent I2SB"] = []
it = iter(loader)
for _ in range(N_BATCHES):
    try:
        a0, a1, ac, am = next(it)
    except StopIteration:
        break
    a0, a1, am, ac = a0.to(device), a1.to(device), am.to(device), ac.to(device)
    yh = _yhat(ac)
    outs = {"UNet (yhat)": yh, "image I2SB": _img_i2sb(a1, yh, ac)}
    if recon_lat is not None:
        outs["latent I2SB"] = latent_i2sb_sample(D_joint, R, D_t1ce, a1, ac, bridge_lat,
                                                 conditional=li.get("conditional", True), M=D_joint.M,
                                                 nfe=NFE, ot_ode=li.get("ot_ode", False),
                                                 decode_dc=li.get("decode_dc", "x1_mean"))
    for name, out in outs.items():
        g, r = apply_loss_mask(a0, out, am, True)
        mets = compute_metrics(g, r, psnr_only=False)
        agg[name].append((float(mets["psnr"]), float(mets["ssim"])))

print(f"{'method':14s} {'PSNR':>8} {'SSIM':>8}   (mean over {sum(len(v) for v in agg.values())//max(len(agg),1)} batches)")
for name, vals in agg.items():
    p = np.mean([v[0] for v in vals]); s = np.mean([v[1] for v in vals])
    print(f"{name:14s} {p:8.2f} {s:8.3f}")
```
