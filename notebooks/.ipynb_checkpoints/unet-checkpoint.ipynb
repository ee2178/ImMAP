```python
import torch
import numpy as np

from diffusion import immap2
from training.common import load_ckpt
from datasets.fastmri.common import load_knee_data
from models import NormUnet
from operators import Mask, Sense, FFT2D
from physics.mask import make_acc_mask
from visualization.image import plot_image
from physics.nle import whiten_from_kspace
```

```python
# GPU Settings
device = 'cuda'
# UNet Checkpoint
unet_path = '/home/ee2178/scratch/ee2178/ImMAP/trained_nets/knee/NormUNet/net.ckpt'
```

```python
kspace, smaps, knee_mask, gnd_truth = load_knee_data(slice_idx=15, device=device)
```

```python
# Mask generation
mask = make_acc_mask(shape=kspace[0,0].shape, accel = 6, acs_lines=20, device = device)
```

```python
def load_model(path, device="cpu"):
    ckpt = torch.load(path, map_location=device)

    model = NormUnet(
        in_chans=2,
        out_chans=2,
        chans=32,
        num_pools=4,
    )

    model.load_state_dict(ckpt["model_state_dict"])
    model.to(device)
    model.eval()

    return model
```

```python
net = load_model(unet_path, device=device)
# net.eval()
```

```python
# Try whitening first
kspace_w, image_w, smaps_w, sigma_n, smaps_w, Zinv = whiten_from_kspace(kspace, smaps, mask = mask, R = 6, gnd_truth_kspace = kspace)
```

```python
# Construct operator
E = Mask(mask) @ FFT2D() @ Sense(smaps_w)
```

```python
# Constructing measurement/simulated kspace
meas_kspace = mask*kspace_w
sim_kspace = E(E.H(kspace_w))
```

```python
zf_recon = E.H(kspace_w)
```

```python
sigma_y = sigma_n
```

```python
recon = immap2(meas_kspace, sigma_y, net, E, lam = 2.0, sigma_max = 1.0, sigma_min = 0.01, save_dir = None, verbose = True, ws = None)
```

```python
plot_image(gnd_truth)
```

```python
plot_image(
    recon*Zinv,
    contrast=False,
    thresh=1.0,
    magnitude=True,
    cmap="gray",
    figsize=None,
    title=None,
    colorbar=False,
)
```
