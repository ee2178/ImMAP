```python
# Construct operator
E = Mask(mask) @ FFT2D() @ Sense(smaps_w)
```

```python
# Constructing measurement/simulated kspace
meas_kspace = mask*kspace_w
sim_kspace = E(E.H(kspace))
```

```python
sigma_y = sigma_n
save_dir = None
```

```python
# Experiment 1: Fixed D operator
from physics.mask import gen_ssdu_mask
# ---------------------------------------------------------
# SSDU Masking Params
# ---------------------------------------------------------
ssdu_masking=False
ssdu_base_accel=1
ssdu_acs=20
ssdu_rho=(0.2, 0.2)
acs_lines = 20
D = Mask(gen_ssdu_mask(gnd_truth[0].shape, acs_lines, ssdu_base_accel, ssdu_acs, ssdu_rho, device = device)) @ FFT2D() @ Sense(torch.ones_like(gnd_truth))
```

```python
M = (gen_ssdu_mask(gnd_truth[0].shape, acs_lines, ssdu_base_accel, ssdu_acs, ssdu_rho, device = device))
```

```python
recon = immap2p5(meas_kspace, sigma_y, net, E,    # Require usual inputs (measurent, nle, net, measurement op)
             sigma_max=1.0,             
             sigma_min=0.01,
             D = D,            # Optional SSDU Operator for Image Domain masking
             organ_mask = None,         # Optional Organ Mask
             ws = None, 
             # Base Diffusion Hyperparameters
             beta = 0.05,    # Noise injection ratio, should belong in [0, 1]
             h_0 = 0.01,      # Initial step size
             save_dir = None,           
             verbose = True, 
             )
```
